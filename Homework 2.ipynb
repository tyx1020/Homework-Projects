> 1.For all interest rates from 1% to 10%, at intervals of 1%, calculate the length of time required for an investment to double in value for each interest rate. *Recall that for an investment that has doubled in size*: 
>
>>$\frac{FV}{PV} = 2$
>>
>>$2 = (1+r)^t$
>>
>>$ln(2) = tln(1+r)$
>>
>>$t = \frac{ln(2)}{ln(1+r)}$
>
>Use the above solution to solve for $t$ for each $r$. Print the solution in a table by using "\n" and "\t" in the >print function when appropriate.
>
>\*Your script *must* use a for loop or for loops to find the solutions.

```python
# Create variables 
Time_dict = {} # initilize a list to trace the time
PV = 1
FV = 1
r = 0.01
margin = 1
t=1
final_FV=2

# create a boolean variable for indicator including true and false 
below_final_FV = True
decrease_margin = False

for i in range(1,11,1):  # r iterate from 1 to 10, r defined in this range                          
    rate = i/100  # r must be floating 
    FV = 1        # every time iterate, reset the value for FV. 
    
    while round(FV,2) != final_FV:
        if decrease_margin == True: 
            margin = margin * .1
            decrease_margin = False
            
        t = t + margin if FV < final_FV else t - margin
        FV = PV * (1 + rate) ** t
        
        if below_final_FV == True and FV > final_FV: 
            decrease_margin = True
            below_final_FV = False
        elif below_final_FV == False and FV < final_FV: 
            decrease_margin = True
            below_final_FV = True
    Time_dict[r] = t
    print(i , ":" , t)
Time_dict
```

```python
#### the first question to initilize a list to trace the time as this is what we are looking for, also create variables 
define FV as 1 every time the for loop selects a new interest rate. 
The final_FV, as the target value for FV, has to be 2 
For loop iterates from 1 to 10, r should not set as 0.01 and it must be floating, the reason why FV is set as 1 is to iterate, 
The difference between while and for is that with the while loop we can execute a set of statements as long as a condition is true.
The concept is that when the FV is less than targeted value and margin is increased, otherwise, it is reduced. 
```

```python
import numpy as np              # needs to define np,NumPy is used to work with arrays
dct = {}
dct["Time to double"] = {}
dct["Rule of 72"] = {}
dct["Rule of 70"] = {}
dct["Computation Solution"] = {}
for r in range (1, 11, 1):
    dct["Time to double"][r] = np.log(2) / np.log(1+r/100)
    dct["Rule of 72"][r] = 72/r 
    dct["Rule of 70"][r] = 70/r 
    PV = 1
    FV = 1
    rate= r/100  
    margin = 1
    t=1
    final_FV=2

# create a boolean variable for indicator including true and false 
    below_final_FV = True
    decrease_margin = False

    rate = r/100 
    FV = 1       
    
    while round(FV,2) != final_FV:
        if decrease_margin == True: 
            margin = margin * .1
            decrease_margin = False
            
        t = t + margin if FV < final_FV else t - margin
        FV = PV * (1 + rate) ** t
        
        if below_final_FV == True and FV > final_FV: 
            decrease_margin = True
            below_final_FV = False
        elif below_final_FV == False and FV < final_FV: 
            decrease_margin = True
            below_final_FV = True
    
    dct["Computation Solution"][r] = t

import pandas as pd
pd.DataFrame(dct)
```
