## 2. Capital Asset Pricing Model 
### 2.1 Overview of CAPM theoretical framework 

CAPM model as one of the most important financial theories, was built on the earlier work on diversification and modern portfolio theory, which was later improved by Sharpe, Markowitz and Merton Miller. The CAPM developed in the 1960s was a true revolution in financial theory, due to their contribution, William F. Sharpe received the Nobel prize in Economic Sciences in 1990 This Model basically is for estimating the capital assets using a linear equilibrium model of returns on investments that explains returns above risk-free rate using covariance with the overall market. (DÅ¾aja & AljinoviÄ‡,2013).

>$E(r_i)\ -\ r_f\ =\ \beta_i(E(r_m) -(r_f))$

where $\beta_i$=$\frac{\sigma_{im}}{\sigma_m^2}$ = $\frac{Cov(r_i,r_m)}{Var(r_{m)}}$ as the coefficient of asset i, $E(r_i)$,$E(r_m)$,r_f, they are respectively expected return rate, expected rate of the effective portfolios, and risk-free return rate for asset i.$E(r_m)$ -$(r_f)$ is called market risk premium, in another words, it is the difference between expected market return rate and risk return rate, this model explains the expected return of one asset is equal to the risk-free interest plus the risk premium (compensation regarding the investors holding relevant investment risks.) , the expected return rate is higher than the risk-free rate. 

Based on the definition, $\beta$ is the ratio between the covariance and market portfolio return. This explains the systematic risk for any kind of capital. CPAM theory says, $\beta$ is a guide to measure the systematic risk of one asset which is to evaluate the whole volatility of one single stock or one portfolio. 

In market equilibrium the CAPM expresses the relationship between an assetâ€™s expected return and so-called systematic risk beta in a liner function. (Sharpe, Litner, Mossin, 1966).In 1972, Fisher Black and Myron Scholes verified the linear correlation between return rate of portfolios and the betas after studying the stock price variation of NYSE from 1931 to 1965. 

Generally speaking, there are two kinds of risk in the financial market, one is called Market risk refers to a risk in a market that cannot be eliminated by diversification. Interest rates, recessions, and wars, for example, are risks that cannot be eliminated by diversification.While, for the unsystematic risk, this includes a new competitor in the market which might drive the sales down from the company invested, or an adjustment with the manaement personnnel, this type of risk can be reduced through diversification.

CAPM predicts that the expected return on an asset equals the risk-free rate plus a risk premium, it is linearly related to systematic risk measured by the assetsâ€™ beta. (Basu and Chawla, 2010).

CAPM has a very simple conclusion: There is only one reason investors get higher returns, and that is by investing in riskier stocks.


Like many scientific models, the CAPM has its drawbacks. In the development of the asset pricing model, the assumptions of the models are( DÅ¾aja AljinoviÄ‡,2013)( Basu& Chawla,2008)( Blac ,Jensen& Scholes,1972).The primary drawbacks are reflected in the model's inputs and assumptions, including:

1)Investors are risk averse and willing to maximize their expected rate of return over a period

2)Investors prefer a portfolio with higher returns 

3)Free to lend and borrow with a risk-free interest 

4)The property is indefinitely divisible 

5)Trade without transaction or taxation costs.

Bescause of these strict assumptions, the CAPM have been questioned such as Eugene Fama and Kenneth French (2004),  Kent Daniel(2001). But due to the simplicity and rationality of CAPM, CAPM still remains very popular in many situations. 


### 2.2 Literature Review 

Numerous empirical tests have been done about CAPM, Fama and MacBeth (1972) did the test showing a positive tradeoff between market portfolio and risk, and average returns reflect the attempts of risk-averse investors to hold efficient portfolios, which are consistent with the modelâ€™s assumptions. After taking the firm size and the book-to market equity into consideration, some researcher like Banz(1981) did the tests showing there was a strong negative relation between average return and firm size, Fama & French(1992) reported in the results, the cross-sectional variation in average stock returns mainly were associated with this two factors.  Pandey and Kok Chee (2000) used the panel data concluding Beta was found to have consistently positive correlation with the stock returns but had weak explanatory power compared with the size. 

But there are more failing empirical tests like DÅ¾aja & AljinoviÄ‡ (2013) who applied this model using monthly stock returns, the results show the CAPM is not adequate for assessing the capital assets on the Central and South- East European emerging securities markets, and the validity of beta is not a valid measure of risk in the experimental markets. Basu and Chawla did the test in India stock market concluding CAPM had weak explanatory power as a suitable descriptor of asset prices in India over the chosen sample. 

The majority tests did by Chinese researchers shows CAPMâ€™s failing explanatory power in stock market, some tests concluded significant correlation( Feng, 2010), some tests show nonlinear correlation(Zhou, 2013), while some tests agree with the CAPM consumptions, Zhu (2010) agreed positive correlation but there was no significant linear correlation with nonsystematic risk. Ye (2010) proved CAPM was not applicable in the A share of Shenzhen stockmarket, he added the variables of market size and average ROE of portfolio and reported they both had significant influence on the expected return. 
 

### 2.3 Methodology 

In empirical analysis, this model is expressed: 


>${R_p=R_f+\beta_p\ (R_m-R_f\ )+\varepsilon_p}$            (1)
        

Where $R_p$ is the expected return of individual or portfolio stock in real estate industry, $ð‘…_ð‘“$ is a risk-free interest rate, $R_m$ represents the expected return of a market portfolio, ${\beta_p}$ measures the systematic risk of individual stock or market portfolio and $\varepsilon_p\$ is the stochastic error term. In this paper, CAPM is used to analyze the Shenzhen property stock market and test whether the effectiveness of this model is verified.

This paper is to use 3 steps method from Black, Jensen and Scholes, which is to time series regression based on CAPM model to make estimation about the individual stock return and market portfolio return. And this will be useful to understand the risk about Shenzhen real estate market. Besides the single index regression, I also use cross sectional regression-based method, FM method (Fama and Macbeth model,MacBeth, 1973) to construct a multiple linear regression model to analyze the combinded effects by the multiple factors, 


The detailed steps are: First, separate the whole time range from May 1st, 2010 to May 1st, 2019 into three periods; Second: Use the data in the first period to calculate the beta for individual stock; Third, rank the individual beta of each stock from lowest to highest, and make different portfolios; Fourth: use the data of second term to estimate the portfolio beta; fifth, make the cross sectional regression using the third period data coupled with these portfolio betas to determine the relationships between expected return and risk. (Tao Zhao, 2001) (Wang and Geng, 2012) (Zhu, 2010).In this paper, we did the smilar way to verify the validity of CPAM in the A share of real estate sector of Shenzhen Stock Exchange.

### 5.3.2 Regression between risk and average return after adding the explainable vaible of residuales

Next, we add the non-system risk to the regression equation to further analyze whether the individual stock is compensated by the market with regards to the nonsystem risk.

>$R_{pi}=\gamma_0+\gamma_1\beta_{pi}+\gamma_2\sigma_{pi}+\varepsilon_i$ (6) 
    
>$\sigma_{pi}$ is the residual obtained from step 2.

#### First run the regression between porpofolio retrun and unsystematic risk

# 6. Conclusion 

In this article, the time series and cross sectional regress was done to test the validity of CAPM about the property A share of Shenzhen stock market, the results show that 

(1) there is no strict positive correlation between the expected return and systematic risk which does not agree with the requirements of CAPM theory; 

(2) $\gamma_0$ is negative and its coefficient value is -0.0005, the deviation from risk-free interest  can be possibly resulted from the opportunistic investment behavior;

(3) the unsystematic risk is necessary to be paid close attention, which might be due to the imperfect stock market system, for example, Chinese securities market and information disclosure system are still immature and needs further to be improved. Therefore, CAPM is not so efficient in Shenzhen real estate stock market over the chosen sample period.


As for the reasons why CAPM is still not applicable in China's Shenzhen A-share real estate market, the reasons might be the following: 

First, the construction of CAPM depends on the various strict assumptions in the actual market, for example, there is no such thing as a perfectly efficient market, a perfectly rational investor with a consistent outlook for the future.

Second, China as a developing country, the establishment of the stock exchange market statered in 1991, compared with other mature stock market like US or UK, it is still new and many laws regulations and regulatory mechanisms still need to improve. 

Though the CAPM is not completely valid in Shenzhen real estate stock market, it still has reference value for the future stock market development, linear correlation between risk and return is the significant requirement for a mature capital market. Even though Shenzhen is undergoing a significant development as a whole, however, in the financial market, it is still not mature as the invertors make investment more out of opportunistic purpose. Property stock market is still very risky and close attention needs to be paid about the rapid fluctuation. More strict policies are necessary to be made to restrain the rapidly increasing of property price and the opportunistic investing behaviors.
